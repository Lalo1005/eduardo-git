```python
# 2. Seleccione los inmuebles clasificados con Tipo 'Casa', 'Casa en condominio' y 'Casa de villa'
# Aquí nos pide 3 criterios de selección, que son 3 condicinales y el operador o representado de esta forma: | .
# Entonces vamos a encapsular entre paréntesis los criterios de selección.
# Entonces estamos indicando que busque la cantidad de locales de Tipo: Casa, Casa en Condominio, Casa de villa.

seleccion = (datos['Tipo'] == 'Casa') | (datos['Tipo'] == 'Casa en condominio') | (datos['Tipo'] == 'Casa de villa')
n2 = datos[seleccion].shape[0]
n2
```

```python
# 3. Seleccione los inmuebles con área entre 60 y 100 mts cuadrados, incluyendo los límites.
# Los límites vendrían a ser >= y <=, si no se incluyesen los límites solo sería > y <.
# Aquí nos pide 2 criterios de selección, que son  condicinales y el operador and representado de esta forma: & .
# Entonces vamos a encapsular entre paréntesis los criterios de selección.
# Entonces estamos indicando que busque los locales con área >=60 y <=100 mts2.

seleccion = (datos['Area']>= 60) & (datos['Area']<= 100)
n3 = datos[seleccion].shape[0]
n3
```

```python
# 4. Seleccione los inmuebles que tengan por lo menos 4 'Cuartos' y 'Alquiler' menor que US$2,000.00

seleccion = (datos['Cuartos']>= 4) & (datos['Valor']< 2000)
n4 = datos[seleccion].shape[0]
n4
```

Siempre se consulta con los jefes o Clientes cuando sea necesario eliminar un registro o varios registros.

***Condición: si el inmueble es un departamento y el mantenimiento es nulo esta línea no la queremos en el dataset.***

Entonces aplicamos un criterio de selección a la base de datos ***datos['Mantenimiento'].isnull()*** datos para obtener de la columnna Mantenimiento cuantos valores NaN existen, pero solo quiero que me muestra la cantidad.

```python
datos[datos['Mantenimiento'].isnull()].shape[0]
```

Hay 2047 líneas nulas en Mantenimiento

Una vez determinados cuantos valores nulos hay en Mantenimiento, creamos un criterio de seleccion para saber cuantos de estos valors nulos son del Tipo Departamento.

El primer criterio es que en la columna Mantenimiento sus valores sean nulos.

```python
seleccion = (datos['Mantenimiento'].isnull()) & (datos['Tipo'] == 'Departamento')
```

```python
seleccion.shape[0]
```

```python
datos.shape[0]
```

* Para saber cuantos registros se van a eliminar.

  Utilizando la virgulilla (~)vamos a traer todo lo contrario a nuestra seleccion.

```python
# Para saber cuantas líneas se van a eliminar.
A = datos.shape[0]
A
# Guardamos como el nuevo contenido del DF datos lo contrario a seleccion.
# Si dejamos datos[seleccion] en realidad vamos a traer los datos que no rqueremos.
# Por eso al colocar la negación delante del criterio de selección retornamos los valores que queremos
datos = datos[~seleccion]
B = datos.shape[0]
B
A - B
```

Se van a eliminar 745 registros. (Líneas, filas).

```python
datos[datos['Mantenimiento'].isnull()].shape[0]
```

Box-plot es un gráfico estadístico que nos permite visualizar los datos de nuestro DF. Utiliza medidas estadísticas para cortar los datos en algunos cuartiles, mediana, máximo y mínimo, e identificar cuales líneas, cuales datos, cuales puntos están fuera de estos cortes.

El Box-plot crea una línea la mediana que divide al 100% de nuestros datos en dos, colocando el 50% en un lado y el otro 50% en otro lado. También crea cuartiles Q1 que selecciona el 25% de los datos, la mediana el 50% y el Q3 que el tercer quartil el 75%, con estas medidas va a crear un límite inferior y un límite superior, donde todos los puntos fuera del límite superior e inferior son nuestros outliers.
Otra medida que existe en el box-plot es el intervalo interquartil que es la resta del Q3 y el Q1.

* Creando un Box-plot

  Es a partir de esta columna que va a realizar la mediana, los quartiles, etc. Tenemos que indicarle cual columna contiene estos datos, no puede ser cualquier tipo de columna tiene que ser una columna que se pueda medir.

* **Eliminando los Outliers.**

```python
# Creando un objeto valor que va tener los datos de la columna 'Valor'
valor = datos['Valor']
valor
```

Y sobre este objeto creamos el primer cuartil, utilizando el método .quantile().

Primer Cuartil = Q1, Tercer cuartil = Q2, intervalo intercuartil = Q3 - Q1 = Q2

```python
Q1 = valor.quantile(0.25)
Q3 = valor.quantile(0.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

El criterio de selección nos va a separar los valores buenos de los valores malos, es decir los registros con información equivocada(que son los 6 registros anteriormente detectados).

Criterio de seleccion: todo lo que sea mayor al limite inferior y
todo lo que sea menor al limite superior es lo que deseo como
resultado final.

```python
# Se tiene que encapsular la operación porque estamos usando el operador &
seleccion = ((valor >= limite_inferior ) & (valor <= limite_superior))
```

Ahora creamos un nuevo DF con la base de datos original y le aplicamos el criterio de seleccion, es decir los que cumplen con el criterio de seleccion. Escogemos lo mejor de la pizza.

```python
datos_new = datos[seleccion]
```

Teniendo esto ya podemos generar un DF excluyendo los outliers según el tipo de cada local.(Este es la parte más interesante).

Para esto creamos un nuevo DF en blanco para crear datos_new en blanco.
Luego vamos a grabar los datos en datos_new, para esto vamos a recorrer cada tipo de mi objeto grupo_tipo y a su propiedad groups y a sus posiciones(keys).

Creamos un criterio de selección con dos variables que verifiquen dos cosas:
(Ambas variables son booleanas)

1.-si es el mismo tipo que esta entrando en el for
   eh_tipo = datos['Tipo'] == tipo
   recorremos toda la columna Tipo y nos preguntamos si es la misma que esta entrando a nuestro for

2.-y si esta dentro de los limites
   eh_dentro_limite = (datos['Valor'] >= limite_inferior[tipo]) & (datos['Valor'] <= limite_superior[tipo])
   Si el valor que entro al nuevo DF es mayor o igual al limite inferior pero de este tipo en particular y si el limite superior del valor sea del mismo tipo. Que esten dentro de los limites del tipo creado para el nuevo DF.

A partir de esas dos variables creamos el criterio de seleccion:
seleccion = eh_tipo & eh_dentro_limite   

Y luego datos seleccion va a seleccionar la fila que cumpla las dos condiciones de la seleccion que acabo de crear.
Y si la seleccion es la correcta quiero que la guarde en datos new

Y para adicionar filas utilizamos el método concat(). le pasamos donde se va a almacenar datos_new y de donte obtengo los datos datos_seleccion.



```python
datos_new = pd.DataFrame()
for tipo in grupo_tipo.groups.keys():
  eh_tipo = datos['Tipo'] == tipo
  eh_dentro_limite = (datos['Valor'] >= limite_inferior[tipo]) & (datos['Valor'] <= limite_superior[tipo])
  seleccion = eh_tipo & eh_dentro_limite
  datos_seleccion = datos[seleccion]
  datos_new = pd.concat([datos_new, datos_seleccion])
```

### **Más sobre gráficos**

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15,10))
```

```python
datos = pd.read_csv('/content/drive/MyDrive/Tratamiento_de_datos/data/alquiler_residencial_sin_outliers.csv', sep=';')
datos.head()
```

```python
# El área es donde va a estar dibujado los gráficos
# Vamos a invocar a nuestra biblioteca matplotlib y al método .figure()
# Esto va a crear una area en nuestro notebook pero todavía no va a mostrar nada.
# Va a ser invocada cuando tenga algo que mostrar
area = plt.figure()
```

Generamos las dimensiones de nuestro gráfico:

* Dimensiones de nuestro gráfico o configurando las posiciones de nuestros gráficos
* add_subplot = adicione un tipo de gráfico a decidir, y dentro de este método
 coloco las coordenadas o posiciones en el que va a estar ubicado el gráfico.
* El gráfico va a estar dividido en 2 lines, 2 columnas, y se ubicará en la
 primera posición.

 1. area.add_subplot(2,2,1) primera posición
 2. area.add_subplot(2,2,2) segunda posición
 3. area.add_subplot(2,2,3) tercera posición
 4. area.add_subplot(2,2,4) cuarta posición

* Por lo que el gráfico va a estar divido en 4.

```python
g1 = area.add_subplot(2,2,1)
g2 = area.add_subplot(2,2,2)
g3 = area.add_subplot(2,2,3)
g4 = area.add_subplot(2,2,4)
```

**Crear gráficos**

1. Gráfico de dispersión o scatter: método .scatter()
2. Histograma: método .hist()
3. Gráfico de líneas: método .plot()
4. Gráfico de barras: método.bar()

```python
# 1. Gráfico de dispersión o scatter: método .scatter()

g1.scatter(datos.Valor, datos.Area)
g1.set_title('Valor por área')

# 2. Histograma: método .hist()

g2.hist(datos.Valor)
g2.set_title('Histograma')

# 3. Gráfico de líneas: método .plot()
# 3.1 Seleccionando una muestra aleatoria con el método .sample()

datos_g3 = datos.Valor.sample(100)

# 3.2 Generando índices en forma secuencial para una visión ordenada

datos_g3.index = range(datos_g3.shape[0])
g3.plot(datos_g3)
g3.set_title('Muestra (Valor)')

# 4. Gráfico de barras: método.bar()
# 4.1 Creando una agrupación

grupo = datos.groupby('Tipo')['Valor']

# 4.2 Generando etiquetas y valores para el gráfico de barras y guardando la media en la variable label y guardando los valores en la variables valores.

label = grupo.mean().index
valores = grupo.mean().values
g4.bar(label, valores)
g3.set_title('Valor Medio por Tipo')
```

Con esto solo configuramos el área:
1. Primero hemos puesto las posiciones.
2. Segundo el tipo de gráfico a visualizar.
