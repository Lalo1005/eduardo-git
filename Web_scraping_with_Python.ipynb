# **1. Web Scraping with Python**

## Instalando el ambiente

```python
from bs4 import BeautifulSoup
```

* Request: son los llamados que vamos a hacer de conexión desde nuestro computador a la página web que queramos.

```python
import requests
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36'}
```

```python
url = "https://www.zoom.com.br/celular"
```

## Comunicándome con la web

Me voy a comunicar con la página y le voy a pedir una respuesta.

```python
respuesta = requests.get(url, headers=headers)
```

Ya me comunique con Amazon voy a visualizar la respuesta.

respuesta que debe ser el código fuentes que acabamos de accesar.

Y nos va a devolver la misma página pero en código.

La siguiente función necesita dos parámetros.
1. **respuesta.content**: el contenido de la respuesta,
y el tipo de estructura como quiero recibirla
2. **lxml**: es un padrón de estructura de página web.

```python
s = BeautifulSoup(respuesta.content, "lxml")
```

```python
print(s)
```

Lo recibido es un texto: que puede ser un texto pequeño o un Big Text.

Para poder trabajarlo y conseguir extraer las palabras llave y la información sustancial que queremos, pues tenemos que utilizar python, utilizar las herramientas del Big data.

Sobre este gran texto vamos a extraer lo que nos interesa los productos y los precios.

# **2. Trabajando la respuesta (Big Text)**

* Ya tenemos la respuesta ahora vamos a trabajarla.

Los resultados quiero guardarlos en dos listas, que va a estar vacías.

```python
prod = []
prec = []
```

Ahora vamos a hacer la magia:
1. Para el producto nos posicionamos encima de la visualización de la imagen y le damos inspeccionar, ubicamos la div que contiene al rótulo título y su clase.

2. Para el precio nos posicionamos encima de la visualización del precio y le damos inspeccionar, ubicamos la div que continen el rótulo precio y su clase.

```python
productos = s.findAll('h2',{'class':'Text_Text__h_AF6 Text_MobileLabelXs__ER_cD Text_DesktopLabelSAtLarge__baj_g ProductCard_ProductCard_Name__LT7hv'})
precios = s.findAll('p',{'class':'Text_Text__h_AF6 Text_MobileHeadingS__Zxam2'})

# ahora vamos a iterar
# el método strip borra los espacios en blanco a la izquierda y a la derecha del resultado.
for item in productos:
  prod.append(item.text.strip())
for item in precios:
  prec.append(item.text.strip())
```

## Ahora para visualizar los datos descargados


* Vamos a iterar y a visualizar uno por uno el producto y su precio.

```python
for i in range(len(prod)):
  print(prod[i], ':' , prec[i])
```
