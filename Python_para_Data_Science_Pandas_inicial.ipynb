## *zip()*

La built in function zip() crea un iterador con tuplas con lo cual se puede combinar listas.

https://docs.python.org/3.6/library/functions.html#zip

````python
carros = ['Jetta Variant', 'Passat', 'Crossfox', 'DS5']
carros
````

````python
valores = [88078.64, 106161.94, 72832.16, 124549.07]
valores
````

````python
zip(carros, valores)
````

el iterador zip fue creado y grabado en <zip at 0x7e0440a48700>

Vamos a convertir este iterador a una lista

````python
list(zip(carros, valores))
````

Se convirtio al iterador zip en una lista pero cada elemento de la lista es una tupla. Y la tupla esta compuesta por el modelo del carro y el valor del mismo. (La magia de python).

No necesito colocar una lista puedo colocar el iterador como parámetro.

````python
for item in zip(carros, valores):
  print(item)
````

Y observamos que cada elemento es una tupla.

Ahora vamos a desempacar tuplas, es decir voy a extraer cada uno de los valores de la tupla y las colocamos dentro de la for.

Ahora en lugar de colocar item voy a desempacar mi tupla.

````python
for carro, valor in zip(carros, valores):
  print(carro, valor)
````

Ahora vemos que se ha separado cada uno de los valores. Y una vez separados puedo hacer cálculos o establecer condiciones.  
Por ejemplo, muestrame los carros con valores mayor a 100,000.00

````python
for carro, valor in zip(carros, valores):
  if(valor > 100000):
    print(carro)
````

También puedo obtener el modelo del carro más su valor.

````python
for carro, valor in zip(carros, valores):
  if(valor > 100000):
    print(carro, valor)
````

# 3.1 Creando diccionarios

**Listas** son colecciones secuenciales, es decir, los elementos de estas secuencias están ordenados y utilizan índices (números enteros) para acceder a los valores.

**Los diccionarios** son colecciones ligeramente diferentes. Son estructuras de datos que representan un tipo de mapeo. ***Los mapeos son colecciones de asociaciones entre pares de valores donde el primer elemento del par se conoce como llave (*key*) y el segundo como valor (*value*).***  

Adicionalmente los diccionarios son mutables.

```
diccionario = {key_1: value_1, key_2: value_2, ..., key_n: value_n}
```

https://docs.python.org/3.6/library/stdtypes.html#typesmapping

*Nota.- cuando quieres accesar a un elemento de una lista tienes que saber su posición para poder ubicarlo y accesar a su valor. En un diccionario no necesitas saber la posición porque los elementos de una diccionario son un par de valores, cada elemento es un par de valores uno ellos es la llave y el otro el valor, entonces para accesar al elemento del diccionario solo necesitas conocer la llave.Si conoces la llave puedes saber el valor de ese elemento*

### Funciones que devuelven más de un valor

#### Formato estándar

```
def <nombre>(<param_1>, <param_2>, ..., <param_n>):
    <instrucciones>
    return (<resultado_1>, <resultado_2>, ..., <resultado_n>)
```

````python
def media(lista):
  valor = sum(lista)/len(lista)
  return(valor, sum(lista), len(lista))
````

````python
media([1,2,3,4,5,6,7,8])
````

## Estructuras de Datos

### Series

Series son arrays unidimensionales etiquetados capaces de almacenar cualquier tipo de dato. Las etiquetas de las líneas se denominan **index**. La forma básica de crear una Series es la siguiente:


```
    s = pd.Series(datos, index = index)
```

El argumento *datos* puede ser un diccionario, una lista, un array Numpy o una constante.

Al crear el pd.Series se visualiza como una tabla con una columna.

### DataFrames

DataFrame es una estructura de datos tabular bidimensional con etiquetas de fila y columna. Como la Series, los DataFrames son capaces de almacenar cualquier tipo de datos.


```
    df = pd.DataFrame(datos, index = index, columns = columns)
```

El argumento *datos* puede ser un diccionario, una lista, un array Numpy, una Series y otro DataFrame.

Al crear el pd.DataFrame se visualiza como una tabla con dos o más columnas.
